```python
novas_cols = {'Unnamed': 'Identificador',
              'name': 'Nome',
              'country': 'País',
              'category': 'Categoria',
              'sales': 'Vendas',
              'profits': 'Lucros',
              'assets': 'Ativos',
              'Valores de Mercado': 'Valores_Mercado'
                            }
```

```python
df_forbes.rename(columns=novas_cols, inplace=True)
df_forbes.head()
```

**Questão 1: Qual é a empresa mais valiosa da lista?**

```python
df_forbes['Valores_Mercado'].max()
```

```python
filtro = df_forbes['Valores_Mercado'] == 328.54
df_forbes[filtro]
```

```python
df_categorias_mais_valiosas = df_forbes.groupby('Categoria').sum('Valores de Mercado')
df_categorias_mais_valiosas.head()
```
